#### How to split by character - character text splitter

This is the simplest method. This splits based on given character sequence, which defaults to "\n\n". Chunk length is measured by number of characters.

1.How thetext is split: by single character seperator.

2.How the chunk size is measured: by nmber of characters.

```python
## Reading a pdf file
from langchain_community.document_loaders import PyMuPDFLoader
pdf_loader=PyMuPDFLoader('VPP.pdf')
docs=pdf_loader.load()
docs
```

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter=CharacterTextSplitter(separator="\n\n",chunk_size=100,chunk_overlap=20)
text_splitter.split_documents(docs)
```
